```python
# Récupération des données historiques de Bitcoin (BTC-USD)

import yfinance as yf
import pandas as pd
import plotly.graph_objects as go

# Définir l'actif et la période
ticker = "BTC-USD"
start_date = "2020-01-01"
end_date = "2025-10-22"

# Télécharger les données
data = yf.download(ticker, start=start_date, end=end_date)

# Réinitialiser les colonnes (éviter les MultiIndex)
data.columns = [col[0] for col in data.columns]

# Afficher les 5 dernières lignes
print(data.tail())

# Sauvegarder les données brutes
data.to_csv('../data/raw/BTC-USD_raw.csv')
print(f"Données sauvegardées")
```

```python
# Visualiser les données de clôture
fig = go.Figure()
fig.add_trace(go.Candlestick(x=data.index, open=data['Open'], high=data['High'], low=data['Low'], close=data['Close'], name='BTC-USD OHLC'))
fig.update_layout( title=f'Analyse du {ticker} ({start_date} à {end_date})', yaxis_title='Prix (USD)', xaxis_title='Date', xaxis_rangeslider_visible=True, template='plotly_dark')
fig.show()
```
